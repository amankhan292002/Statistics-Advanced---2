1. What is hypothesis testing in statistics?
     - Hypothesis testing is a statistical method used to make decisions or draw conclusions about a population based on sample data. It involves formulating assumptions (hypotheses) and using statistical techniques to determine whether there is enough evidence to support or reject these assumptions.

     ->Key Components of Hypothesis Testing

       Null hypothesis (H₀) - the default assumption

       Alternative hypothesis (H₁)- the claim being tested

       Test statistic - calculated from sample data

       Significance level (α) - threshold for decision making

       p-value - probability of observing the result under H₀

     ->Purpose

       To validate research claims

       To compare population parameters

       To support data-driven decision making

     ->Example

       Testing whether the average marks of students differ from a known standard.




2.  What is the null hypothesis, and how does it differ from the alternative
hypothesis?
    - The null hypothesis states that there is no effect, no difference, or no relationship between variables. It is the assumption that any observed difference is due to random chance

  -> Example:
      
      H₀: The average sales are equal to 100 units.

   ->Alternative Hypothesis (H₁ or Hₐ)

     The alternative hypothesis states that there is an effect, difference, or relationship. It contradicts the null hypothesis and represents the research claim.

     -> Example:
         
         H₁: The average sales are not equal to 100 units.

  ->Key Differences
       Null Hypothesis (H₀)	Alternative Hypothesis (H₁)

       Assumes no change       	Assumes change exists

      Tested directly         	Supported indirectly

      Accepted or rejected	     Accepted if H₀ is rejected

3.  Explain the significance level in hypothesis testing and its role in deciding the outcome of a test.
    - The significance level (α) is the maximum probability of rejecting the null hypothesis when it is actually true. It represents the risk of making a Type I error.

    -> Common Significance Levels
       
        α = 0.05 (5%)

        α = 0.01 (1%)

        α = 0.10 (10%)

    -> Role in Decision Making
       
        If p-value ≤ α → reject the null hypothesis

        If p-value > α → fail to reject the null hypothesis

    -> Importance
        
        Controls false positive results

        Determines test strictness

        Balances risk and confidence

    ->Example

        At α = 0.05, we accept a 5% chance of rejecting a true null hypothesis.



4. What are Type I and Type II errors? Give examples of each.
     - Type I Error
         
           A Type I error occurs when the null hypothesis is rejected even though it is actually true. It is also known as a false positive.

           Probability of Type I error = α (significance level)

      ->Example:
          
           A medical test concludes that a person has a disease when the person is actually healthy.


      -> Type II Error

            A Type II error occurs when the null hypothesis is not rejected even though it is false. It is also known as a false negative.

            Probability of Type II error = β

        -> Example:

             A medical test fails to detect a disease in a person who actually has the disease.


        

5. What is the difference between a Z-test and a T-test? Explain when to use
each.
     - Both Z-test and T-test are statistical tests used to compare sample data with population parameters, but they are used under different conditions.

     -> Z-test
          
          A Z-test is used when:
          
          Population standard deviation (σ) is known

          Sample size is large (n ≥ 30)

          Data follows a normal distribution

      ->Example

           Testing whether the average height of students differs from a known population mean.

      -> T-test

           A T-test is used when:
            
           Population standard deviation is unknown

           Sample size is small (n < 30)

           Data is approximately normally distributed

        -> Example:
            
            Comparing the average marks of a small group of students with a standard value.

```python
#7.
# # Implement hypothesis testing using Z-statistics for a sample dataset in
# Python. Show the Python code and interpret the results.
# sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
# 50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
# 50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
# 50.3, 50.4, 50.0, 49.7, 50.5, 49.9]
# (Include your Python code and output in the code box below.)


# We perform a Z-test to test whether the sample mean is significantly different from a known population mean.

#Null Hypothesis (H₀): μ = 50
#Alternative Hypothesis (H₁): μ ≠ 50
#Significance level α = 0.05


sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]


import numpy as np
from scipy import stats

sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

# Sample statistics
sample_mean = np.mean(sample_data)
sample_std = np.std(sample_data, ddof=1)
n = len(sample_data)

# Population mean
mu = 50

# Z-statistic
z_stat = (sample_mean - mu) / (sample_std / np.sqrt(n))

# P-value (two-tailed test)
p_value = 2 * (1 - stats.norm.cdf(abs(z_stat)))

print("Sample Mean:", sample_mean)
print("Z-statistic:", z_stat)
print("P-value:", p_value)
```

```python
#8.
# Write a Python script to simulate data from a normal distribution and
# calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.
# (Include your Python code and output in the code box below.)


# #To simulate data from a normal distribution, we generate random values using NumPy by specifying the mean and standard deviation.
import numpy as np
import matplotlib.pyplot as plt
import math

# Simulate normal distribution
np.random.seed(42)
data = np.random.normal(loc=50, scale=10, size=1000)

# Sample statistics
mean_val = np.mean(data)
std_val = np.std(data, ddof=1)
n = len(data)

# 95% confidence interval
z = 1.96
margin_error = z * (std_val / math.sqrt(n))
ci_lower = mean_val - margin_error
ci_upper = mean_val + margin_error

print("Mean:", mean_val)
print("95% Confidence Interval:", (ci_lower, ci_upper))

# Plot histogram
plt.hist(data, bins=30, edgecolor='black')
plt.axvline(mean_val, color='red', linestyle='--', label='Mean')
plt.title("Histogram of Simulated Normal Data")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```

```python
#9.
#  Write a Python function to calculate the Z-scores from a dataset and
# visualize the standardized data using a histogram. Explain what the Z-scores represent
# in terms of standard deviations from the mean.
# (Include your Python code and output in the code box below.)

import numpy as np
import matplotlib.pyplot as plt

def calculate_z_scores(data):
    mean = np.mean(data)
    std = np.std(data, ddof=1)
    z_scores = [(x - mean) / std for x in data]
    return z_scores

# Dataset
data = [45, 50, 52, 48, 47, 53, 55, 49, 51, 54]

# Calculate Z-scores
z_scores = calculate_z_scores(data)

print("Z-scores:")
print(z_scores)

# Plot histogram
plt.hist(z_scores, bins=10, edgecolor='black')
plt.xlabel("Z-score")
plt.ylabel("Frequency")
plt.title("Histogram of Z-scores")
plt.show()
```
